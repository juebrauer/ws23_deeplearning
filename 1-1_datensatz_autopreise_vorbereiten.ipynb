# Alles Bisherige in einer Codezelle

```python
# 1. Datensatz runterladen
download = False
if download:
    import urllib.request
    fname = "data_cars.csv"
    URL = "http://www.juergenbrauer.org/datasets/" + fname
    urllib.request.urlretrieve(URL, fname)

# 2. Einlesen des Datensatzes
import pandas
t = pandas.read_csv("data_cars.csv")

# 3. Unbenötigte Spalten verwerfen
t.drop("Unnamed: 0", axis="columns", inplace=True)

# 4. Alle unvollständigen Zeilen verwerfen
t.dropna(inplace=True)

# 5. Preis-Spalte enthält falsche Werte, d.h. teilweise keine Zahlen
#    Diese verwerfen.
spalten = ["price_in_euro", "year", "power_kw", "power_ps"]
for spalte in spalten:
    t[spalte] = pandas.to_numeric(t[spalte], errors="coerce")
    t.dropna(inplace=True)

# 6. Entferne l/100km bzw. g/km in den fuel_consumption... Spalten
t["fuel_consumption_l_100km"] = t["fuel_consumption_l_100km"].str.replace(",", ".").str.extract("(\d+\.\d+)")
t["fuel_consumption_g_km"]    = t["fuel_consumption_g_km"].str.extract("(\d+)")

t["fuel_consumption_l_100km"] = t["fuel_consumption_l_100km"].astype(float)
t["fuel_consumption_g_km"]    = t["fuel_consumption_g_km"].astype(float)
t.dropna(inplace=True)

# 7. Registration Date Spalte aufdröseln in Monate + Jahre
t["registration_month"] = pandas.to_datetime(t["registration_date"], format="%m/%Y").dt.month
t["registration_year"] = pandas.to_datetime(t["registration_date"], format="%m/%Y").dt.year
t.drop("registration_date", axis="columns", inplace=True)
```

```python
# 1. Datensatz runterladen
download = False
if download:
    import urllib.request
    fname = "data_cars.csv"
    URL = "http://www.juergenbrauer.org/datasets/" + fname
    urllib.request.urlretrieve(URL, fname)

# 2. Einlesen des Datensatzes
import pandas
t = pandas.read_csv("data_cars.csv")

# 3. Unbenötigte Spalten verwerfen
t.drop( ["Unnamed: 0", "offer_description"], axis="columns", inplace=True)

# 4. Alle unvollständigen Zeilen verwerfen
t.dropna(inplace=True)

# 5. Preis-Spalte enthält falsche Werte, d.h. teilweise keine Zahlen
#    Diese verwerfen.
spalten = ["price_in_euro", "year", "power_kw", "power_ps"]
for spalte in spalten:
    t[spalte] = pandas.to_numeric(t[spalte], errors="coerce")
    t.dropna(inplace=True)

# 6. Entferne l/100km bzw. g/km in den fuel_consumption... Spalten
t["fuel_consumption_l_100km"] = t["fuel_consumption_l_100km"].str.replace(",", ".").str.extract("(\d+\.\d+)")
t["fuel_consumption_g_km"]    = t["fuel_consumption_g_km"].str.extract("(\d+)")

t["fuel_consumption_l_100km"] = t["fuel_consumption_l_100km"].astype(float)
t["fuel_consumption_g_km"]    = t["fuel_consumption_g_km"].astype(float)
t.dropna(inplace=True)

# 7. Registration Date Spalte aufdröseln in Monate + Jahre
t["registration_month"] = pandas.to_datetime(t["registration_date"], format="%m/%Y").dt.month
t["registration_year"] = pandas.to_datetime(t["registration_date"], format="%m/%Y").dt.year
t.drop("registration_date", axis="columns", inplace=True)

# 8. Ausreißer (heute: Extremwerte bzgl. km-Stand und Preis) rausfiltern
p95_km = t["mileage_in_km"].quantile(0.95)
p95_price = t["price_in_euro"].quantile(0.95)
t = t.query(f"price_in_euro < {p95_price} and mileage_in_km < {p95_km}")

# 9. registration_date und year speichern die gleichen Werte
#    Deshalb einer der beiden Spalten verwerfen
t = t.drop("year", axis="columns")

# 10. Bereinigte Tabelle speichern
t.to_csv("data_cars_cleaned.csv", index=False)
```
